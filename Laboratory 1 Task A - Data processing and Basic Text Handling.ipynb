```python
import nltk
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import brown
from nltk.probability import FreqDist
```

```python
nltk.download("brown")
```

```python
brownWords = brown.words()
brownSentences = brown.sents()
```

```python
#calculate frequency of each word and save them in a text file
#here we will find out frequency of each word and we will save it in an txt file named "brownWordFrequency"
frequencyOfWord = FreqDist(brownWords)
with open("brownWordFrequency.txt", "w") as file:
    for word, frequency in frequencyOfWord.items():
        file.write(f"{word}: {frequency}\n")
```

```python
#calculate top=30 frequent words
#Here we will take two empty list x, y. we will iterate top 30 most common value of frequencyofword dictionary and assign it on x and y respectively and then we will plot it on matplotlib
x=[]
y=[]
for word, frequency in frequencyOfWord.most_common(30):
    x.append(word)
    y.append(frequency)
    print(f"{word}: {frequency}")
plt.figure(figsize=(12, 6))
plt.bar(x,y)
plt.title("Top 30 commo words in brown corpus")
plt.xlabel('word')
plt.ylabel('frequency')
plt.show()
```

```python
#calculate last=30 frequent word
#Here we will take two empty list x, y. we will iterate most common value of frequencyofword dictionary but from the last side and assign it on x and y respectively. By this we will get the last 30 common word. and then we will plot it on matplotlib

leastFrequentWord = frequencyOfWord.most_common()[-30:]
x=[]
y=[]
for word, frequency in leastFrequentWord:
    x.append(word)
    y.append(frequency)
    print(f"{word}: {frequency}")
plt.figure(figsize=(20, 10))
plt.barh(x,y)
plt.title("Last 30 common words in brown corpus")
plt.xlabel('word')
plt.ylabel('frequency')
plt.show()
```

```python
#calculate last=30 frequent word
#step1: calculate the length of number of words
#step2: divide the length by 2
# now we will thake the most common words range from middle length-15 to middle length+15
# so it will give us most common words from middle range

x=[]
y=[]

totalNumberOfWords = len(frequencyOfWord)
middleFrequentWord = frequencyOfWord.most_common()[(totalNumberOfWords//2)-15:(totalNumberOfWords//2)+15]
for word, frequency in middleFrequentWord:
    x.append(word)
    y.append(frequency)
    print(f"{word}: {frequency}")

plt.figure(figsize=(20, 10))
plt.barh(x,y)
plt.title("30 common words in brown corpus from Middle")
plt.xlabel('word')
plt.ylabel('frequency')
plt.show()
```

```python
# find out same length of words from the dataset:
# step1: create a dictionary that will satore: key= length of word, value = how many times that length of word occurs
# step2: iterate through all words and calculate the length of the word
# step3: if same length of key is already exist in our dictionary than increase the value by 1
# step4: else assign the new length key in our dictionay with the value 1
# step5: plot the key value pair on the histogram

sameLengthOfWords = {}
for word in brownWords:
    wordLength = len(word)
    
    if wordLength in sameLengthOfWords:
        sameLengthOfWords[wordLength] +=1
    else:
        sameLengthOfWords[wordLength]=1

x=[]
y=[]
for sameLength, frequency in sameLengthOfWords.items():
    x.append(sameLength)
    y.append(frequency)
    print(f"Word Length= {sameLength}: {frequency}")
    

plt.figure(figsize=(20, 10))
plt.bar(x,y)
plt.xlabel('length')
plt.ylabel('frequency')
plt.show()
```

```python
#calculate the frequency of model word and word length and character length of the sentence:

# step1: create a list of model word and a dictionary that indicates how many time a model word occurs in the datasets, key will be each word from the model words list
# step2: iterate through all model words list and assign the value of the key through .count(word) methods of NLTK. so we will get how many times each model words occurs into the datasets
# step3: To calculate the word lehngth and character length of the sentence . frist we will take all the sentence seperately through NLTK. sent() method
# step4: Now we will iterate through all the sentence and if model words exist in the sentence we will count number of words and number of character of the sentence 
# step5: than we will assign the value of word length and character length to the new two list respectively

modelWords = ["will", "must", "might", "may", "could", "can"]
numberOfModelWords = {}

for word in modelWords:
    numberOfModelWords[word] = brownWords.count(word)
    
for word, frequency in numberOfModelWords.items():
    print(f"{word} = {frequency}")
    

brownSentence = brown.sents()
wordLength = []
characterLength = []

for sent in brownSentence:
    for word in sent:
        if word.lower() in modelWords:
            wordLength.append(len(sent))
            characterLength.append(len(" ".join(sent)))

print("Length of words where model words found")
print(wordLength)
print("Length of character where model words found")
print(characterLength)
```

```python
#import stopwords for english
from nltk.corpus import stopwords
nltk.download("stopwords")
stopWords = set(stopwords.words("english"))
```

```python
#calculate the number of stopwords in the sentence and frequency of stopwords with the word length and character length of the sentence:
# step1: import stopwords from nltk and set it with english stopwords
# step2: create 3 list. one for number of stopwords in the sentence and other two to calculate the word length and character length in the stopwords sentence
# step3: iterate through all sentence and compare whether the sentence has any stopwords or not 
# step4: If the sentence has any stopwords. calculate the number number of stop words and the words length and characther lehgth of the sentence. and assign them to their list
# step5: Now we will make two plot to show how frequent stopwords occuers with respect to words and character

numberOfStopWords = []
numberOfWords = []
numberOfCharacter = []

for sentence in brownSentence:
    lowerSentence = [word.lower() for word in sentence]
    stopWordsInSentence = [word for word in lowerSentence if word in stopWords]
    if stopWordsInSentence:
        numberOfStopWords.append(len(stopWordsInSentence))
        numberOfWords.append(len(sentence))
        numberOfCharacter.append(len(" ".join(sentence)))
        
plt.figure(figsize=(10, 5))
plt.hist2d(numberOfWords, numberOfStopWords, cmap=plt.cm.Blues)
plt.colorbar(label="Frequency")
plt.title("Frequency of Stopwords vs. Sentence Length in Words")
plt.xlabel("Sentence Length (in Words)")
plt.ylabel("Number of Stopwords")
plt.show()


plt.figure(figsize=(10, 5))
plt.hist2d(numberOfCharacter, numberOfStopWords, cmap=plt.cm.Purples)
plt.colorbar(label="Frequency")
plt.title("Frequency of Stopwords vs. Sentence Length Characters")
plt.xlabel("Sentence Length (in Characters)")
plt.ylabel("Number of Stopwords")
plt.show()
```
